## Binomial Distribution

Similar to the above problem, this time again we randomly sample 20 fruits.

**What is the probability that this time the sample contains 5 apples and 15 oranges?**

Please note that this time the order of the fruits being picked does not matter.

````python
# your solution here
def combinacion(n, k):
    return math.factorial(n) // (math.factorial(k) * math.factorial(n - k))
````

````python
print(combinacion(60, 5))
````

````python

print(combinacion(40, 15))
````

````python
print(combinacion(100, 20))
````

````python
(combinacion(60, 5))*(combinacion(40, 15)) / (combinacion(100, 20))
````

In the lesson we also discussed CDF (cumulative distribution function). In the context of the above problem, **what is the probability that less than 5 fruits picked are apples?**

Write the formula and calculate the probability below.

````python
# your code here

n = 20
p = 0.6

cdf = sum(math.comb(n, k) * (p**k) * ((1-p)**(n-k)) for k in range(5))
print(cdf)
````

**Plot the Probability Mass Function (PMF) for the above binomial distribution.**

*Hint: The range of x in a binomial distribution is from `0` to `n`.*

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters of the binomial distribution
n = 10      # number of trials
p = 0.5     # probability of success in each trial

# Range of possible values (0 to n)
x = np.arange(0, n+1)

# PMF: probability of each outcome
pmf = binom.pmf(x, n, p)
````

````python
# your code here
# Please label the axes and give a title to the plot 

plt.figure(figsize=(10, 6))
plt.bar(x, pmf, color='steelblue', edgecolor='black')
plt.xlabel('Successes')
plt.ylabel('Probability')
plt.title('Mass Funciont Probability')
plt.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
````
